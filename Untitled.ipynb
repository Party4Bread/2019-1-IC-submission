## PREPARE STAGE

```python
from datetime import datetime,timedelta,date
from pathlib import Path
from pytz import timezone, utc
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.cluster import DBSCAN
from sklearn.svm import OneClassSVM
print("Load OK")

p=Path("../0608data/data")
KST = timezone('Asia/Seoul')
print("Preset OK")
```

## IMPORT SINGLE DATASET

```python
target='ST-1121'
fromdate=datetime(2019,4,1)
todate=datetime(2019,6,1)

todate+=timedelta(days=1)#for chk
fromdate=KST.localize(fromdate)
todate=KST.localize(todate)

dataz=[]
curdataz=[]
curdst=fromdate+timedelta(days=1)

for file in p.glob('*'):
    jk=file.stem
    bt=datetime.strptime(jk,"%Y%m%d %H_%M_%S")
    kt=utc.localize(bt).astimezone(KST)
    if curdst<kt:
        curdst+=timedelta(days=1)
        curdataz=[]
        if curdst>todate:
            break
        
    if fromdate<=kt<curdst:
        with file.open(encoding='utf8') as k:
            data=json.load(k)
        for row in data:
            if row["stationId"]==target:
                curdataz.append(int(row["parkingBikeTotCnt"]))
                break                
        dataz.append(curdataz)
    
len(dataz),len(dataz[0])
```

### 하루 시작 시점부터 누적 변화량 평균

```python
%matplotlib inline
tk=np.array([0]*144)
cnt=0
for i in dataz:
    if len(i)!=144:
        continue
    cnt+=1
    hk=np.array(i)
    hk-=hk[0]
    tk+=hk

plt.plot(tk/cnt)
plt.show()
```

### 10분당 순간 변화량 평균

```python
%matplotlib inline
tk=np.array([0]*144)
cnt=0
for i in dataz:
    if len(i)!=144:
        continue
    cnt+=1
    hk=np.array(i)
    hk-=hk[0]
    for i in reversed(range(1,144)):
        hk[i]-=hk[i-1]
    tk+=hk

plt.plot(tk/cnt)
plt.show()
```

## 기간 데이터 로드후 nv에 저장

```python
fromdate=datetime(2019,4,1)
todate=datetime(2019,6,1)

todate+=timedelta(days=1)#for chk
fromdate=KST.localize(fromdate)
todate=KST.localize(todate)

dataz=[]
curdataz=[]
curdst=fromdate+timedelta(days=1)

allstation=[]
allloc=[]
alltot=[]
allval={}

#find all target
for file in p.glob('*'):
    jk=file.stem
    bt=datetime.strptime(jk,"%Y%m%d %H_%M_%S")
    kt=utc.localize(bt).astimezone(KST)
    
    if fromdate<=kt:
        with file.open(encoding='utf8') as k:
            data=json.load(k)
        for row in data:
            allstation.append(row["stationId"])
            allloc.append((float(row["stationLatitude"]),float(row["stationLongitude"])))
            alltot.append(int(row["rackTotCnt"]))
        break
print("Target Found")
#print(len(allstation))

#crawl all data
curval={}
for ik in allstation:
    allval[ik]=[]
    curval[ik]=[]
for file in p.glob('*'):
    jk=file.stem
    bt=datetime.strptime(jk,"%Y%m%d %H_%M_%S")
    kt=utc.localize(bt).astimezone(KST)
    if curdst<kt:
        curdst+=timedelta(days=1)
        if curdst>todate:
            break
        curval={}
        for ik in allstation:
            curval[ik]=[]
            
    if fromdate<=kt<curdst:
        with file.open(encoding='utf8') as k:
            data=json.load(k)
        for row in data:
            if row["stationId"] in allstation:
                curval[row["stationId"]].append(int(row["parkingBikeTotCnt"]))
        for ik in allstation:
            allval[ik].append(curval[ik])
print('Preloading...')

numval=[]
for ik in allstation:
    tk=np.array([0]*144)
    cnt=0
    for i in allval[ik]:
        if len(i)==0:
            continue
        while len(i)<144:
            i.append(i[-1])
        cnt+=1
        hk=np.array(i)
        hk-=hk[0]
        tk+=hk
    if cnt>0:
        numval.append(tk/cnt)
    else:
        numval.append(tk)
        print(ik)
nv=np.array(numval)
#svdata={'as':allstation,'al':allloc,'at':alltot,'av':allval}
#json.dump(svdata,open('temp.json','w+',encoding='utf8'),ensure_ascii=True)
print('Done')
```

```python
print(int((len(numval)/2)**0.5))
print(int((len(numval))**0.5))
nv=np.array(numval)

kmeans = KMeans(n_clusters=26,max_iter=100000,n_init=100)
kmeans.fit(nv)
```

```python
X=np.array(allloc)

plt.figure(figsize=(10, 10))
plt.scatter(X[:, 1], X[:, 0], c=kmeans.labels_)
plt.xlim([126.78,127.19])
plt.ylim([37.43,37.7])

plt.show()
```

```python
clustersz=26
fig, axs = plt.subplots(clustersz,1,figsize=(15,clustersz*8))
for i in range(clustersz):
    li=np.where(kmeans.labels_ == i)[0]
    for j in li:
        axs[i].plot(nv[j])
plt.show()
```

```python
clf = IsolationForest()
out=clf.fit_predict(nv)
ot=np.where(out == -1)[0]
print(len(ot))
noot=[]
for i in range(len(nv)):
    if i not in ot:
        noot.append(nv[i])

nootv=np.array(noot)

clustersz=int((len(numval)/2)**0.5)
km = KMeans(n_clusters=clustersz,max_iter=100000,n_init=100)
kres = km.fit_predict(nootv)

fig, axs = plt.subplots(clustersz,1,figsize=(15,clustersz*8))
for i in range(clustersz):
    li=np.where(kres == i)[0]
    for j in li:
        axs[i].plot(nootv[j])
plt.show()
```

```python
hNk=[]
for i,j in enumerate(nv):
    hNk.append(j/alltot[i])
hNk=np.array(hNk)

clf = IsolationForest()
ont=clf.fit_predict(hNk)
on=np.where(ont == -1)[0]
print(len(on))

noht=[]
for i in range(len(hNk)):
    if i not in on:
        noht.append(hNk[i])

nohtv=np.array(noht)

clustersz=5#int((len(nohtv)/2)**0.5)
km = KMeans(n_clusters=clustersz,max_iter=100000,n_init=100)
kres = km.fit_predict(nohtv)

fig, axs = plt.subplots(clustersz,1,figsize=(10,clustersz*3),dpi=200)
for i in range(clustersz):
    axs[i].set_title(f"class {i}")
    li=np.where(kres == i)[0]
    axs[i].set_ylim([-1.5,1.5])
    for j in li:
        axs[i].plot(nohtv[j])
plt.show()
fig.savefig('bikestationclass.png')

```

```python
fig, axs = plt.subplots(clustersz,1,figsize=(10,clustersz*3))

for i in range(clustersz):
    axs[i].set_title(f"class {i}")
    li=np.where(kres == i)[0]
    axs[i].set_ylim([-1.0,1.0])
    avg=np.array([0.0]*144)
    for j in li:
        avg+=nohtv[j]
    avg/=len(li)
    axs[i].plot(avg)
plt.show()
fig.savefig('bikestationclassavg.png')
```
