```python
import csv
cl=[]

for i in range(5):
    cl.append([0]*5)
    
with open('subwayclass.csv', 'r', encoding='utf-8') as k:
    with open('bikeclass.csv', 'r', encoding='utf-8') as f:
        sr = csv.reader(k)
        br = csv.reader(f)
        gosu=[]
        for j in br:
            gosu.append(j)
        
        lili=[len(list(filter(lambda a:int(a[-1])==x,gosu))) for x in range(5)]
        
        for i in sr:
            if i[2]=='100':
                continue
            for j in gosu:
                try:
                    if (float(i[3])-float(j[1]))**2+(float(i[4])-float(j[2]))**2<0.00001:#0.0000077:
                        if j[4]=='100':
                            continue
                        cl[int(i[2])][int(j[4])]+=1
                except:
                    pass
        
print(lili)             
import matplotlib.pyplot as plt
import numpy as np
fig, axs = plt.subplots(5,1,figsize=(10,5*3))
cl=np.array(cl)/np.array(lili)
for i in range(5):
    axs[i].set_title(f"subway class {i}")
    axs[i].bar(range(5),cl[i])
plt.show()
fig.savefig('neargap.png')
```
